```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
data = pd.read_csv('ghealth.csv')
data.head()
```

```python
data.groupby(['Category'])[['Urban', 'Rural']].sum()
```

```python
data.groupby(['Category'])[['Urban', 'Rural']].sum().plot(kind = 'bar', figsize = (10, 10))
```

# Inferences
- In rural areas, Maternal and Child Health Maternity Care is High
- In both Urban and Rural areas, there is no awareness on 'Screening for cancer among women (age 30-49 years)'
- Blood sugar level is very low in urban and rural areas especially in women

```python

data[data['Category'] == 'Screening for Cancer among Women (age 30-49 years)'][(list(data)[3:])].plot(kind = 'bar', figsize = (30, 10))
```

```python
data.groupby(['Category'])[['Hyderabad']].sum().plot(kind = 'bar', figsize = (10, 10))
```
